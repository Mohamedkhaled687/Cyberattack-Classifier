# Introduction

- The dataset used in this project is the UNSW-NB15 dataset, a modern benchmark for evaluating network intrusion detection systems (NIDS). It was created at the Cyber Range Lab of UNSW Canberra using the IXIA PerfectStorm tool to simulate a mix of real-world normal network activities and contemporary synthetic attack behaviors

# Data Type & Description

### Desription

- The UNSW-NB15 consists of network traffic records . These Features capture a wide range of information about each network connection , including basic connection details , content-based attributes , time-based statistics , and traffic flow characteristics, The dataset is designed to provide a comprehensive view of both normal and malicious network activities, making it suitable for developing and evaluating intrusion detection systems.


#### Feature Groups :
- Basic Features : Include source and destination IP addresses and ports , protocal type , and connection state.

- Content Features : Capture Information about the content of the network , such as number of bytes and packets transfeered and service types.

- Time Features : Measure timing aspects of the connection such as duration , inter-packet arrival times , and jitter

- Additional Generated Features : Include statisitical summaries and counts of connections sharing similar properties within short time window .

```python
# Handle outliers in df_train using IQR method for numeric features

numeric_features = df_train.select_dtypes(include=[np.number]).columns.tolist()

for col in numeric_features:
    Q1 = df_train[col].quantile(0.25)
    Q3 = df_train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the outliers
    df_train[col] = np.where(df_train[col] < lower_bound, lower_bound,
                             np.where(df_train[col] > upper_bound, upper_bound, df_train[col]))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
categorical_cols = X_train.select_dtypes(include=['object']).columns

for col in categorical_cols :
    X_train[col] = encoder.fit_transform(X_train[col])
    X_test[col] = X_test[col].astype(str).apply(lambda val: val if val in encoder.classes_ else 'unknown')
    encoder.classes_ = np.append(encoder.classes_, 'unknown')  # add 'unknown' to known classes
    X_test[col] = encoder.transform(X_test[col])


y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)

```
